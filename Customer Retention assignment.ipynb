# 1. Distribution of Male and Female customers who buy online :

# 2. Distribution of male and female buyers agewise:

```python
sns.countplot("2 How old are you? ", hue="1Gender of respondent", data=df)
```

From the above plot, 1 in age stands for less than 20 years, 2 stands for 21-30 years, 3 stands for 31-40 years, 4 stands for 41-50 years and finally 5 stands for 51 years and above. Therefore, it can be interpreted that females of the age-group 21-30 years are more into online shopping followed by the females of the age-group 31-40 years. The least buyers are the males of the age-group 51 and above. Hence it can be concluded that, online shopping is more-youth centric. 

# 3. Citywise inclination of online shopping :

```python
plt.figure(figsize=(30,6))
sns.countplot(x="3 Which city do you shop online from?", data=df)
plt.title("Cities from where customers shop online")
plt.show()
```

From the above plot, it can be observed that Delhi is the city from where we can get most of our online shoppers and especially the females of the age-group 21-30. Delhi followed by Greater Noida, Banglore, Noida , Karnal and Ghaziabad are the cities of most of the online shoppers.

# 12. Analyzing the customer satisfaction through shopping online and what impacts it.

```python
#subplot 1
fig = plt.figure(figsize=(10,8))
plt.subplot()
plt.title('subplot: Quality Information')
sns.countplot(data = dataset, x = '35 Displaying quality Information on the website improves satisfaction of customers')

#subplot 2
fig = plt.figure(figsize=(10,8))
plt.subplot()
plt.title('subplot:Shopping through a Good platform ')
sns.countplot(data = dataset, x = "36 User derive satisfaction while shopping on a good quality website or application")

#subplot 3
fig = plt.figure(figsize=(10,8))
plt.subplot()
plt.title('subplot:Users satisfaction')
sns.countplot(data = dataset, x = "37 Net Benefit derived from shopping online can lead to users satisfaction")


#subplot 4
fig = plt.figure(figsize=(10,8))
plt.subplot()
plt.title('subplot:Need of trust in User Satisfaction')
sns.countplot(data = dataset, x = "38 User satisfaction cannot exist without trust")
```

Customers Strongly agree that user satisfaction is a very important factor that websites should also take into consideration. From, the above plot, it is clear that factors like : clear description of information on products , shopping from a qualified online website etc. leads to give users the satisfaction of shopping as well as trust is build in this process.
